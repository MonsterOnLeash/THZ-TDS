```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import *
import itertools
from itertools import combinations
from hapi import *

db_begin('data')
%matplotlib inline
```

```python
fetch('NO', 8, 1,  10, 100)
fetch('N2O', 4, 1,  10, 100)
fetch('NH3', 11, 1,  10, 100)
fetch('CO', 5, 1,  10, 100)
```

```python
def mixture(compnum, comp, press, temp, props):
    dil = {c:p for c, p in list(zip(comp, props))[1:]}
    dil['self'] = props[0]
    nu, coef = absorptionCoefficient_Voigt(Components=compnum, SourceTables=comp, OmegaRange=(10, 100), OmegaStep=0.01, HITRAN_units=False, Environment={'p':press,'T':temp}, Diluent=dil)
    return nu, coef
```

```python
def single(compnum, comp, press, temp):
    nu, coef = absorptionCoefficient_Voigt(Components=compnum, SourceTables=comp, OmegaRange=(10, 100), OmegaStep=0.01, HITRAN_units=False, Environment={'p':press,'T':temp}, GammaL='gamma_self')
    return nu, coef
```

```python
isos = [(8,1), (4,1), (11,1),(5,1)]
comps = ['NO', 'N2O', 'NH3', 'CO']
```

```python
res = []
cond = []
parts = []
components = []
```

```python
isos1 = tuple(itertools.permutations(isos, 1))
comps1 = tuple(itertools.permutations(comps, 1))
isos3 = tuple(itertools.permutations(isos, 3))
comps3 = tuple(itertools.permutations(comps, 3))
```

```python
for t in range(286, 307, 10):
    for p in np.arange(0.5, 1.6, 0.5):
        for c in zip(comps1, isos1):
            n, au = single(c[1], c[0][0], p, t)
            res.append(au)
            cond.append((p, t))
            components.append(c[1][0])
```

```python
for t in range(286, 307, 10):
    for p in np.arange(0.5, 1.6, 0.5):
        for c in zip(comps3, isos3):
            for i in np.arange(0.5, 1, 0.1):
                for j in np.arange(0.1, 1-i-0.005, 0.1):
                    n, au = mixture(c[1], c[0], p, t, (i, j, 1-i-j))
                    res.append(au)
                    cond.append((p, t))
                    parts.append((1))
                    components.append(c[1])
```

```python
data = pd.DataFrame(res)
```

```python
ts = [i[1] for i in cond]
ps = [i[0] for i in cond]
```

```python
data['t'] = ts
data['p'] = ps
```
